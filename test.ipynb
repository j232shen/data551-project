```python
import pandas as pd
import altair as alt
```

```python
wfp = pd.read_parquet('data551-project/data/processed/wfp_preprocessed.parquet')
wfp.head()
```

```python
# Count the number of unique countries for each commodity
commodity_country_count = (
    wfp.groupby('commodity')['country'].nunique()  # Count unique countries per commodity
    .reset_index(name='country_count')  # Reset index to have a clean DataFrame
)

# Sort commodities by the number of countries they appear in
sorted_commodities = commodity_country_count.sort_values(by='country_count', ascending=False)
sorted_commodities = sorted_commodities.reset_index(drop=True)

# Check the result
sorted_commodities.head(10)
```
